```python
# Core Budget Logic
def calculate_budget(income, housing, food, transport, other expenses):
    total = housing + food + transport + other expenses
    savings = income - total
    
    if savings < 0:
        status = "Over Budget"
    elif savings < income * 0.2:
        status = "Low Savings"
    else:
        status = "Healthy Savings"
    
    return total, savings, status
```
